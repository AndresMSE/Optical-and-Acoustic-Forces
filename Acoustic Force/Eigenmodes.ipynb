```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eigenv = pd.read_csv('eigenvectors.csv',header=None)
modelmesh = pd.read_csv('modelmesh.csv',header=None)
```

```python
modelmesh
```

```python
eigenv.T
```

```python
nump = eigenv.T.to_numpy()
```

```python
mesh = modelmesh.T.to_numpy()
```

```python
meshco = []
for cord in mesh:
    meshco.append([cord[0],cord[1]])
```

```python
x,y = np.zeros(len(mesh)),np.zeros(len(mesh))
for i in range(len(mesh)):
    x[i] = meshco[i][0]
    y[i] = meshco[i][1]
```

```python
zvals = []
for mode in range(len(nump)):
    z = np.zeros(len(mesh))
    for k in range(len(nump[mode])):
        z[k] = nump[mode][k]
    zvals.append(z)
    
```

```python
np.gradient(zvals[0])
```

```python
# %matplotlib qt
%matplotlib inline
def plt_mode(m):
    %matplotlib qt
    color_map = plt.get_cmap('seismic')
    fig = plt.figure(figsize=(20,20))
    ax = fig.gca(projection='3d')
    preassure = ax.scatter3D(x, y, (np.gradient(zvals[m]))**2, c= (zvals[m])**2,cmap = color_map)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('Presión [au]')
    ax.set_title(f'Eigenmodo #{m}')
    plt.colorbar(preassure)
    plt.show()
```

```python
plt_mode(6)
```

```python
%matplotlib qt
def plot_modes():
    %matplotlib inline
    color_map = plt.get_cmap('seismic')
    no_modes = int(input('¿Cuántos modos propios deseas calcular?'))
    if no_modes%3 ==0:
        scale = no_modes//3
    else:
        if no_modes%2 ==0:
            scale = no_modes//2
        else:
            scale = no_modes%3 +1
    if no_modes<3:
        length = no_modes
    else:
        length = 3
    pixels = 100*(min(length,scale))
    print(scale)
    if int(no_modes) > len(zvals):
        print('Excede el número de modos')
    else:

        fig = plt.figure(figsize=(5*length,15*scale),dpi=pixels)

        for j in range(int(no_modes)):
            ax = fig.add_subplot(int(no_modes), 3, j+1, projection='3d')
            preassure = ax.scatter3D(x, y, zvals[j], c= zvals[j],cmap = color_map)
            ax.set_xlabel('x')
            ax.set_ylabel('y')
            ax.set_zlabel('Presión [Pa]')
            ax.set_title(f'Eigenmodo {j+1}')
    fig.colorbar(preassure,ax =ax,orientation='vertical')
```

```python
plot_modes()
```

```python
def pressure_x(x,xlist,zval,m):
    inde = np.where(xlist==x)
    
    return zval[m][inde[0][0]]
    
```
